<span style="font-family: Arial; font-weight:bold;font-size:1.2em;color:#aec101">Bar(count) plot for checking the Embarked vs (No. of passengers, sex, Survived , Pclass). And your Observations

```python
# create subplot plot
f,ax=plt.subplots(2,2,figsize=(20,10))

# create Bar (count) plot for Embarked vs. No. Of Passengers Boarded
sns.countplot(x='Embarked',data = df,ax=ax[0,0],color = 'skyblue')
ax[0,0].set_title('No. Of Passengers Boarded for Embarked')

# create Bar (count) plot for Embarked vs. Male-Female Split
sns.countplot(x = 'Embarked',data = df,hue = 'Sex',ax = ax[0,1], color = 'g')
ax[0,1].set_title('Male Female Split for Embarked')

# create Bar (count) plot for Embarked vs Survived
sns.countplot(x = 'Embarked',data = df,hue = 'Survived',ax = ax[1,0], color = 'teal')
ax[1,0].set_title('Survived vs non Survived for Embarked')

# create Bar (count) plot for Embarked vs Pclass
sns.countplot('Embarked',hue='Pclass',data=data,ax=ax[1,1],color = 'darkblue')
ax[1,1].set_title('Embarked vs Pclass')
plt.subplots_adjust(wspace=0.2,hspace=0.5)
plt.show()
```
